```python
import time
from pathlib import Path
import instaloader
import pandas as pd


def get_followed_users(username, password, save_path):
    save_path = Path(save_path)
    # Create the directory if it does not exist
    save_path.mkdir(parents=True, exist_ok=True)

    # Instantiate Instaloader with desired options
    L = instaloader.Instaloader(
        download_pictures=True,
        download_videos=False,
        download_video_thumbnails=False,
        compress_json=False,
        download_geotags=False,
        post_metadata_txt_pattern=None,
        max_connection_attempts=0,
        download_comments=False,
    )

    # Login
    L.context.log("Logging in...")
    try:
        L.load_session_from_file(username)
    except FileNotFoundError:
        L.context.log("Session file does not exist yet - Logging in.")
    if not L.context.is_logged_in:
        L.context.log("Login required.")
        L.login(username, password)
        L.save_session_to_file()

    # Get the profile of the specified user
    profile = instaloader.Profile.from_username(L.context, username)

    # Print number of followed profiles to scrape
    followees_count = profile.followees
    L.context.log(f"Number of followed profiles to scrape: {followees_count}")

    # Initialize metadata table
    metadata = []

    # Loop through all followees
    for person in profile.get_followees():
        try:
            username = person.username
            username_path = save_path / username

            # Create the directory for the user's posts if it does not exist
            username_path.mkdir(parents=True, exist_ok=True)

            # Get the posts for the user
            posts = person.get_posts()
            
            # Loop through the last n user's posts
            for i, post in enumerate(posts):
                 # Download the post and save to the user's directory, if it does not already exist
                if not L.download_post(post, username_path):
                    print(f"Post already exists, skipping waiting time")
                    continue
                try:
                    
                    # Add metadata to table
                    metadata.append({
                        "username": username,
                        "image_name": post.date_utc.strftime("%Y-%m-%d_%H-%M-%S"),
                        "image_path": str(username_path / f"{post.date_utc.strftime('%Y-%m-%d_%H-%M-%S')}.jpg"),
                        "image_description": post.caption,
                        "image_link": f"https://www.instagram.com/p/{post.shortcode}/",
                    })
                    
                    # Download the post and save to the user's directory
                    L.context.log(f"Downloading post {post.shortcode}")
                    L.download_post(post, username_path)
                    
                    post_sleep = 5  # Sleep 5 seconds between posts
                    L.context.log(f"Sleeping for {post_sleep} seconds")
                    time.sleep(post_sleep)


                except instaloader.exceptions.TooManyRequestsException as e:
                    L.context.log(f"Too many queries in the last time. Need to wait {e.retry_after} seconds.")
                    time.sleep(e.retry_after + 10)

                except Exception as e:
                    L.context.log(f"Error downloading post {post.shortcode}: {e}")

            L.context.log(f"Username '{username}' scraped!")

        except Exception as e:
            L.context.log(f"Error scraping user '{username}': {e}")

    # Save metadata to Excel file
    df = pd.DataFrame(metadata)
    metadata_path = save_path / "metadata.xlsx"
    df.to_excel(metadata_path, index=False)
    L.context.log(f"Metadata saved to {metadata_path}")
```
